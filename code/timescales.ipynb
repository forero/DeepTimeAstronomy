```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.constants as c
import astropy.units as u
%matplotlib inline
```

```python
mass = np.logspace(-4,15,10) * u.solMass
long = np.logspace(-6,3,5) * u.parsec
print(mass)
```

```python
for l in long:
    time = np.sqrt(1.0/(c.G * (mass/l**3))).to(u.year)
    #print(time)
    plt.plot(np.log10(mass/u.solMass), np.log10(time/u.year), label='${:.2E} L/pc=$'.format((l/u.parsec)))
plt.ylim([-2,5])
plt.xlabel('$log_{10}$ Mass / Msun')
plt.ylabel('$log_{10}$ Time / year')
plt.legend()
#useful for galaxy dynamics
```

```python
velocity = np.logspace(0,4,10) * u.km/u.s
time = np.logspace(1,4,5) * u.year
```

```python
for t in time:
    d = (velocity * t).to(u.parsec)
    plt.plot(np.log10(velocity/(u.km/u.s)), np.log10(d/u.parsec), label='{:.2E}'.format(t))
    #print(time)
plt.ylim([-6,2])
plt.xlabel('$log_{10}$ Velocity / km/s')
plt.ylabel('$log_{10}$ Distance / parsec')
plt.legend()
# Useful for high velocity stars in our galaxy
```
